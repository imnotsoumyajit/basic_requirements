```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('Data/housing.csv')
df.head()
```

```python
rooms=df[['total_bedrooms','total_rooms']]
rooms
```

```python
plt.figure(dpi=75)
plt.hist(df['total_rooms'])

plt.title('Histogram of Rooms')
plt.ylabel('Count')
plt.xlabel('Price')
```

```python
plt.figure(dpi=75)
plt.plot(df['total_bedrooms'][:10000])
```

```python
plt.figure(dpi=75)
plt.scatter(df['total_bedrooms'][:100],df['total_rooms'][:100])
```

```python
rooms.corr()
```

```python
plt.figure(dpi=75)
plt.scatter(df['total_bedrooms'][:100],df['total_rooms'][:100])
plt.scatter(df['total_bedrooms'][100:200],df['total_rooms'][100:200])
plt.scatter(df['total_bedrooms'][200:230],df['total_rooms'][200:230])
plt.legend(['First 100','Second 100','Third 30'])
```

```python
plt.boxplot(df['total_rooms'])
```

```python
bins_series=pd.cut(df['total_rooms'],bins=4).value_counts()
bins_series
```

```python
values=bins_series.to_list()
values
```

```python
labels=[str(x) for x in bins_series.index.to_list()]
labels
```

```python
plt.figure(dpi=80)
plt.pie(values,labels=labels)
```

```python
plt.bar(labels,height=values)
```
